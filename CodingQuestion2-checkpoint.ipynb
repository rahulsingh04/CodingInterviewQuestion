```python
def unique_permutation(nums):
    if len(nums) == 0: return [[]]  # Base case: single empty permutation
    
    nums.sort()  # Sort to handle duplicates

    AllPermutation = []
    
    # Go through each element and recursively generate permutations
    for i in range(len(nums)):
        # Skip duplicates: if the current element is the same as the previous one, continue
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        
        # Get the remaining elements
        remaining = nums[:i] + nums[i+1:]
        
        #      
        for permutation in unique_permutation(remaining):
            AllPermutation.append([nums[i]] + permutation)
    
    return AllPermutation
```

```python
nums = [1, 1, 2]
unique_permutation(nums)
```

```python
def isSubsequence(s, t):
    i,j = 0,0
    s_len = len(s)
    t_len = len(t)
    while i<s_len and j<t_len :
        if s[i] == t[j] :
            i+=1
        j+=1
    return i == len(s)
```

```python
s = "axc"
t = "ahbgdc"
isSubsequence(s, t)
```

```python
def Max_No_Of_K_SumPair(nums, k):
    result_list = [] 
    for num in nums:
        current_num = num 
        comp_num =  k - num 
        if comp_num in nums:
            result_list.append([current_num, comp_num])
            nums.remove(current_num)
            # nums.remove(comp_num)
    my_list = [list(item) for item in set(tuple(x) for x in result_list)]

    print("Result List Is :->", result_list)
    print("After Removing Duplicates :->", my_list)
    print("Currents nums list is :->", nums)
```

```python
nums = [3,1,3,4,3]
k = 6
Max_No_Of_K_SumPair(nums, k)
```

```python
def spiral_matrix(n):
    # Create matrix of n*n filled with zero 
    matrix = [[0]*n for _ in range(n)]
    top, bottom, left, right = 0, n-1, 0, n-1
    num = 1
    ## Filling The Matrix in Spiral Order 

    while top <= bottom and left <= right : 
        # Filling The Top Row 
        for i in range(left, right+1):
            matrix[top][i] = num
            num+=1
        top+=1

        # Fillng The Right Columns 
        for i in range(top, bottom+1):
            matrix[i][right] = num 
            num+=1
        right-=1

        # Filling The Bottom Row 
        if top <= bottom:
            for i in range(right, left-1, -1):
                matrix[bottom][i] = num
                num+=1
            bottom-=1

        ## Filling The Left Columns :
        if left <= right :
            for i in range(bottom, top-1, -1):
                matrix[i][left] = num
                num+=1

            left+=1
        
    return matrix
```

```python
print(spiral_matrix(4))
```
